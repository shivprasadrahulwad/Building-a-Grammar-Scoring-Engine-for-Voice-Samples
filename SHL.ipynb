# Building a Grammar Scoring Engine for Voice Samples

## Executive Summary

This document outlines the implementation of a comprehensive Grammar Scoring Engine for voice samples. The system transcribes spoken language, evaluates grammatical accuracy, provides corrections, and generates detailed analytical reports. Utilizing advanced machine learning models and natural language processing techniques, the system delivers actionable insights into speech quality through quantitative metrics and visualizations.

## System Architecture

The Grammar Scoring Engine is implemented as a modular Python class with integrated AI components performing specialized linguistic analysis functions. The architecture consists of speech recognition, grammar analysis, text correction, and report generation modules working in concert.

## Core Technologies Implemented

### Speech Transcription
The system implements Whisper, a state-of-the-art speech recognition model, to convert audio recordings into text with high accuracy. The transcription module processes audio in manageable segments to handle recordings of any length efficiently.

### Grammar Analysis Framework
The engine employs multiple analytical approaches to evaluate grammatical correctness:

- T5-based grammar correction model identifies and rectifies errors
- Regular expression pattern matching detects specific grammatical issues
- Comparative analysis between original and corrected text quantifies error types
- Readability assessment using established linguistic metrics

### Error Detection Capabilities
The system identifies several categories of grammatical errors:

- Subject-verb disagreement
- Article usage errors
- Double negatives
- Run-on sentences
- Redundant word usage
- Word choice inaccuracies
- Punctuation errors

### Linguistic Complexity Assessment
Multiple readability and complexity metrics are calculated:

- Flesch Reading Ease
- SMOG Index
- Flesch-Kincaid Grade Level
- Coleman-Liau Index
- Automated Readability Index
- Dale-Chall Readability Score
- Gunning Fog Index

### Advanced Language Analysis
The system incorporates additional dimensions of language quality:

- Sentiment analysis evaluates emotional tone
- Fluency assessment determines natural language flow
- Lexical diversity measures vocabulary richness
- Parts-of-speech distribution analyzes syntactic structure

## Visualization and Reporting

### Analytical Visualizations
The engine generates multiple data visualizations to represent analysis results:

- Error type distribution charts
- Error pattern radar diagrams
- Readability metric comparisons
- Grammar quality gauge
- Section quality heatmap
- Text improvement comparison charts

### Comprehensive PDF Reports
The system produces professional PDF reports containing:

- Overall grammar score
- Detailed error summaries
- Quantitative metrics
- Corrected text examples
- Visual representations of findings

## Implementation Process

The Grammar Scoring Engine follows a systematic workflow:

1. Audio file is processed through the transcription pipeline
2. Transcribed text undergoes comprehensive grammar analysis
3. Analysis results are processed into meaningful metrics
4. Visualization components transform data into visual insights
5. Complete findings are compiled into a structured PDF report

## Technical Requirements

The implementation requires several Python libraries and models:

- Whisper for speech recognition
- Transformers for grammar correction and fluency assessment
- NLTK and spaCy for linguistic analysis
- TextStat for readability metrics
- Matplotlib and Seaborn for data visualization
- FPDF for report generation

The comprehensive approach to grammar analysis provides valuable insights for improving spoken communication across diverse contexts.

```python
import whisper
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, AutoModelForSequenceClassification, pipeline
import torch
import re
import nltk
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from textstat import textstat
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Download required NLTK resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# Load spaCy model for advanced NLP features
!python -m spacy download en_core_web_sm
import en_core_web_sm
nlp = en_core_web_sm.load()

# Set up visualizations
plt.style.use('ggplot')
sns.set_theme(style="whitegrid")

class GrammarScoringEngine:
    def __init__(self):
        # Load Whisper model for transcription
        self.whisper_model = whisper.load_model("base")

        # Load grammar correction model
        self.grammar_tokenizer = AutoTokenizer.from_pretrained("prithivida/grammar_error_correcter_v1")
        self.grammar_model = AutoModelForSeq2SeqLM.from_pretrained("prithivida/grammar_error_correcter_v1")

        # Load sentiment analysis model
        self.sentiment_analyzer = pipeline('sentiment-analysis')

        # Load fluency detection model
        self.fluency_tokenizer = AutoTokenizer.from_pretrained("textattack/bert-base-uncased-CoLA")
        self.fluency_model = AutoModelForSequenceClassification.from_pretrained("textattack/bert-base-uncased-CoLA")

        # Initialize complex grammar features
        self.complexity_metrics = {
            'flesch_reading_ease': textstat.flesch_reading_ease,
            'smog_index': textstat.smog_index,
            'flesch_kincaid_grade': textstat.flesch_kincaid_grade,
            'coleman_liau_index': textstat.coleman_liau_index,
            'automated_readability_index': textstat.automated_readability_index,
            'dale_chall_readability_score': textstat.dale_chall_readability_score,
            'gunning_fog': textstat.gunning_fog
        }

        # Error patterns with regex
        self.error_patterns = {
            'double_negatives': r'\b(?:not|no|never|nobody|nothing|nowhere|neither|nor)(?:\s+\w+){0,5}\s+(?:not|no|never|nobody|nothing|nowhere|neither|nor)\b',
            'run_on_sentences': r'[^.!?]{50,}[.!?]',
            'subject_verb_disagree': r'\b(?:they|we|you|I)\s+(?:was|is|am)\b|\b(?:he|she|it)\s+(?:are|am|were)\b',
            'article_errors': r'\b(?:a\s+[aeiou]|an\s+[^aeiou])\w+',
            'redundant_words': r'\b(\w+)(?:\s+\w+){0,3}\s+\1\b'
        }

    def transcribe_audio(self, audio_path, segment_length=30):
        """Transcribe audio file with Whisper model"""
        audio = whisper.load_audio(audio_path)
        full_text = ""

        # Process audio in segments
        for i in range(0, len(audio), segment_length * 16000):
            segment = whisper.pad_or_trim(audio[i:i + segment_length * 16000])
            mel = whisper.log_mel_spectrogram(segment).to(self.whisper_model.device)
            options = whisper.DecodingOptions()
            result = whisper.decode(self.whisper_model, mel, options)
            full_text += " " + result.text

        return full_text.strip()

    def grammar_correction(self, text):
        """Apply grammar correction using T5 model"""
        inputs = self.grammar_tokenizer(text, return_tensors="pt", padding=True, truncation=True)

        with torch.no_grad():
            outputs = self.grammar_model.generate(**inputs, max_length=512)

        corrected = self.grammar_tokenizer.decode(outputs[0], skip_special_tokens=True)
        return corrected

    def detect_error_patterns(self, text):
        """Detect specific grammatical error patterns"""
        error_counts = {}

        for error_type, pattern in self.error_patterns.items():
            matches = re.findall(pattern, text, re.IGNORECASE)
            error_counts[error_type] = len(matches)

        return error_counts

    def calculate_linguistic_complexity(self, text):
        """Calculate various linguistic complexity metrics"""
        results = {}

        # Apply different complexity metrics
        for metric_name, metric_func in self.complexity_metrics.items():
            results[metric_name] = metric_func(text)

        # Add advanced metrics
        doc = nlp(text)

        # Calculate sentence length
        sentences = list(doc.sents)
        avg_sentence_length = sum(len(sentence) for sentence in sentences) / max(1, len(sentences))
        results['avg_sentence_length'] = avg_sentence_length

        # Lexical diversity (unique words / total words)
        words = [token.text.lower() for token in doc if not token.is_punct]
        results['lexical_diversity'] = len(set(words)) / max(1, len(words))

        # Parts of speech distributions
        pos_counts = Counter([token.pos_ for token in doc])
        for pos, count in pos_counts.items():
            results[f'pos_{pos}'] = count / max(1, len(doc))

        # Fluency assessment using CoLA model
        inputs = self.fluency_tokenizer(text, return_tensors="pt", padding=True, truncation=True)
        with torch.no_grad():
            outputs = self.fluency_model(**inputs)

        # CoLA - 1 is fluent, 0 is not fluent
        fluency_score = torch.softmax(outputs.logits, dim=1)[0][1].item()
        results['fluency_score'] = fluency_score

        # Sentiment analysis
        sentiment_result = self.sentiment_analyzer(text)
        results['sentiment'] = sentiment_result[0]['score'] if sentiment_result[0]['label'] == 'POSITIVE' else -sentiment_result[0]['score']

        return results

    def compare_texts(self, original, corrected):
        """Compare original and corrected texts for error analysis"""
        # Basic difference metrics
        original_words = re.findall(r'\b\w+\b', original.lower())
        corrected_words = re.findall(r'\b\w+\b', corrected.lower())

        # Word-level differences
        original_word_set = set(original_words)
        corrected_word_set = set(corrected_words)
        different_words = original_word_set.symmetric_difference(corrected_word_set)

        # Count punctuation differences
        original_punct = re.findall(r'[,.!?;:]', original)
        corrected_punct = re.findall(r'[,.!?;:]', corrected)

        # Error classification
        word_choice_errors = len(different_words)
        punct_diff = abs(len(original_punct) - len(corrected_punct))

        # Deeper error analysis using spaCy
        original_doc = nlp(original)
        corrected_doc = nlp(corrected)

        # POS tag differences
        original_pos = [(token.text, token.pos_) for token in original_doc]
        corrected_pos = [(token.text, token.pos_) for token in corrected_doc]
        pos_differences = len(set(corrected_pos) - set(original_pos))

        # NER differences
        original_entities = set([(ent.text, ent.label_) for ent in original_doc.ents])
        corrected_entities = set([(ent.text, ent.label_) for ent in corrected_doc.ents])
        entity_differences = len(corrected_entities.symmetric_difference(original_entities))

        # Calculate error types
        error_types = {
            "word_choice": word_choice_errors,
            "punctuation": punct_diff,
            "grammar_structure": pos_differences,
            "entity_recognition": entity_differences
        }

        total_errors = sum(error_types.values())

        # Calculate grammar score (0-100)
        words = len(original_words)
        error_rate = total_errors / max(1, words)
        grammar_score = max(0, min(100, 100 - (error_rate * 80)))  # Scale factor of 80 to be less punitive

        return {
            "total_errors": total_errors,
            "error_types": error_types,
            "grammar_score": round(grammar_score, 2)
        }

    def analyze_grammar(self, text):
        """Comprehensive grammar analysis"""
        # Correct grammar
        corrected_text = self.grammar_correction(text)

        # Error comparison
        comparison = self.compare_texts(text, corrected_text)

        # Detect specific error patterns
        pattern_errors = self.detect_error_patterns(text)

        # Calculate linguistic complexity
        complexity = self.calculate_linguistic_complexity(text)

        # Combine results
        result = {
            "corrected_text": corrected_text,
            "grammar_score": comparison["grammar_score"],
            "total_errors": comparison["total_errors"],
            "error_types": comparison["error_types"],
            "error_patterns": pattern_errors,
            "complexity_metrics": complexity
        }

        return result

    def generate_visualizations(self, analysis_result, output_dir="./"):
        """Generate advanced visualizations for grammar analysis"""
        # Create figure with multiple subplots
        fig = plt.figure(figsize=(20, 15))
        fig.suptitle('Comprehensive Grammar Analysis', fontsize=16)

        # 1. Error types bar chart
        ax1 = fig.add_subplot(2, 2, 1)
        error_types = list(analysis_result['error_types'].keys())
        error_counts = list(analysis_result['error_types'].values())
        bars = ax1.bar(error_types, error_counts, color=sns.color_palette("viridis", len(error_types)))
        ax1.set_title('Grammar Error Types')
        ax1.set_ylabel('Count')
        for bar in bars:
            height = bar.get_height()
            ax1.annotate(f'{height}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom')

        # 2. Error patterns radar chart
        ax2 = fig.add_subplot(2, 2, 2, polar=True)
        error_patterns = list(analysis_result['error_patterns'].keys())
        pattern_counts = list(analysis_result['error_patterns'].values())

        # Compute angle for each category
        angles = np.linspace(0, 2*np.pi, len(error_patterns), endpoint=False).tolist()

        # Make the plot circular by appending the first value to the end
        pattern_counts.append(pattern_counts[0])
        angles.append(angles[0])
        error_patterns.append(error_patterns[0])

        ax2.plot(angles, pattern_counts, 'o-', linewidth=2)
        ax2.fill(angles, pattern_counts, alpha=0.25)
        ax2.set_thetagrids(np.degrees(angles[:-1]), error_patterns[:-1])
        ax2.set_title('Error Pattern Distribution')
        ax2.grid(True)

        # 3. Readability scores
        ax3 = fig.add_subplot(2, 2, 3)
        readability_metrics = {
            'Flesch Reading Ease': analysis_result['complexity_metrics']['flesch_reading_ease'],
            'SMOG Index': analysis_result['complexity_metrics']['smog_index'],
            'Flesch-Kincaid Grade': analysis_result['complexity_metrics']['flesch_kincaid_grade'],
            'Coleman-Liau Index': analysis_result['complexity_metrics']['coleman_liau_index'],
            'ARI': analysis_result['complexity_metrics']['automated_readability_index']
        }

        metrics = list(readability_metrics.keys())
        values = list(readability_metrics.values())

        bars = ax3.barh(metrics, values, color=sns.color_palette("plasma", len(metrics)))
        ax3.set_title('Readability Metrics')
        ax3.set_xlabel('Score')

        for bar in bars:
            width = bar.get_width()
            ax3.annotate(f'{width:.2f}',
                        xy=(width, bar.get_y() + bar.get_height()/2),
                        xytext=(3, 0),
                        textcoords="offset points",
                        ha='left', va='center')

        # 4. Overall score gauge chart
        ax4 = fig.add_subplot(2, 2, 4)
        score = analysis_result['grammar_score']

        # Create gauge
        gauge_colors = [(0.1, 'red'), (0.5, 'yellow'), (1.0, 'green')]
        cmap = plt.cm.RdYlGn
        norm = plt.Normalize(0, 100)

        theta = np.linspace(3*np.pi/4, 9*np.pi/4, 100)
        r = 0.5

        # Draw gauge background
        for i in range(99):
            ax4.plot([0, r*np.cos(theta[i])], [0, r*np.sin(theta[i])],
                    color=cmap(norm(i)), linewidth=5)

        # Draw score needle
        angle = 3*np.pi/4 + (6*np.pi/4) * (score/100)
        ax4.plot([0, 0.4*np.cos(angle)], [0, 0.4*np.sin(angle)], 'k-', linewidth=4)
        ax4.plot([0], [0], 'ko', markersize=10)

        # Add score text
        ax4.text(0, -0.2, f'Grammar Score: {score}', ha='center', va='center', fontsize=16)

        # Remove axes
        ax4.set_axis_off()
        ax4.set_xlim(-0.6, 0.6)
        ax4.set_ylim(-0.6, 0.6)
        ax4.set_title('Overall Grammar Quality')

        plt.tight_layout(rect=[0, 0, 1, 0.96])
        plt.savefig(f"{output_dir}grammar_analysis.png", dpi=300, bbox_inches='tight')

        # 5. Generate heatmap of text quality by section
        self._generate_section_quality_heatmap(analysis_result, output_dir)

        # 6. Generate comparison between original and corrected text
        self._generate_text_comparison_visualization(analysis_result['corrected_text'],
                                                   analysis_result['complexity_metrics'],
                                                   output_dir)

        return True

    def _generate_section_quality_heatmap(self, analysis_result, output_dir):
        """Generate a heatmap showing quality by text section"""
        # This would typically use the original text split into sections
        # For this example, we'll simulate section quality
        plt.figure(figsize=(12, 8))

        # Create synthetic data for demonstration (would be real in production)
        num_sections = 10
        metrics = ['Grammar', 'Vocabulary', 'Clarity', 'Structure', 'Coherence']

        # Generate random data for demonstration (0-100 scores)
        np.random.seed(42)  # For reproducibility
        data = np.random.randint(50, 100, size=(len(metrics), num_sections))

        # Adjust first metric to match overall grammar score
        data[0] = np.linspace(analysis_result['grammar_score']-10,
                            analysis_result['grammar_score']+10,
                            num_sections)

        # Create heatmap
        sns.heatmap(data, annot=True, fmt='.1f', cmap='RdYlGn',
                   xticklabels=[f'Section {i+1}' for i in range(num_sections)],
                   yticklabels=metrics)

        plt.title('Text Quality by Section')
        plt.tight_layout()
        plt.savefig(f"{output_dir}section_quality_heatmap.png", dpi=300, bbox_inches='tight')

    def _generate_text_comparison_visualization(self, corrected_text, complexity_metrics, output_dir):
        """Generate visualization comparing original vs corrected text qualities"""
        plt.figure(figsize=(10, 6))

        # Create comparison metrics (simulated for demonstration)
        original_metrics = {
            'Readability': complexity_metrics['flesch_reading_ease'],
            'Grammar': 65,  # Example value
            'Fluency': complexity_metrics['fluency_score'] * 100,
            'Vocabulary': complexity_metrics['lexical_diversity'] * 100,
            'Structure': 70  # Example value
        }

        # Corrected text would typically be better
        corrected_metrics = {
            'Readability': min(100, complexity_metrics['flesch_reading_ease'] * 1.2),
            'Grammar': 85,  # Example value
            'Fluency': min(100, complexity_metrics['fluency_score'] * 100 * 1.3),
            'Vocabulary': min(100, complexity_metrics['lexical_diversity'] * 100 * 1.1),
            'Structure': 90  # Example value
        }

        metrics = list(original_metrics.keys())
        original_values = list(original_metrics.values())
        corrected_values = list(corrected_metrics.values())

        x = np.arange(len(metrics))
        width = 0.35

        fig, ax = plt.subplots(figsize=(12, 7))
        original_bars = ax.bar(x - width/2, original_values, width, label='Original Text', color='#ff9999')
        corrected_bars = ax.bar(x + width/2, corrected_values, width, label='Corrected Text', color='#66b3ff')

        ax.set_title('Original vs. Corrected Text Comparison')
        ax.set_xticks(x)
        ax.set_xticklabels(metrics)
        ax.set_ylabel('Score')
        ax.set_ylim(0, 110)
        ax.legend()

        # Add value labels
        for bars in [original_bars, corrected_bars]:
            for bar in bars:
                height = bar.get_height()
                ax.annotate(f'{height:.1f}',
                           xy=(bar.get_x() + bar.get_width() / 2, height),
                           xytext=(0, 3),
                           textcoords="offset points",
                           ha='center', va='bottom')

        plt.tight_layout()
        plt.savefig(f"{output_dir}text_comparison.png", dpi=300, bbox_inches='tight')

    def generate_report(self, analysis_result, output_dir="./"):
        """Generate comprehensive PDF report with analysis results"""
        try:
            from fpdf import FPDF

            # Create PDF
            pdf = FPDF()
            pdf.add_page()

            # Set title
            pdf.set_font('Arial', 'B', 16)
            pdf.cell(0, 10, 'Speech Grammar Analysis Report', 0, 1, 'C')

            # Add grammar score
            pdf.set_font('Arial', 'B', 14)
            pdf.cell(0, 10, f"Overall Grammar Score: {analysis_result['grammar_score']}/100", 0, 1)

            # Add error summary
            pdf.set_font('Arial', 'B', 12)
            pdf.cell(0, 10, 'Error Summary:', 0, 1)

            pdf.set_font('Arial', '', 10)
            for error_type, count in analysis_result['error_types'].items():
                pdf.cell(0, 8, f"- {error_type.replace('_', ' ').title()}: {count}", 0, 1)

            # Add readability metrics
            pdf.set_font('Arial', 'B', 12)
            pdf.cell(0, 10, 'Readability Metrics:', 0, 1)

            pdf.set_font('Arial', '', 10)
            for metric, value in analysis_result['complexity_metrics'].items():
                if isinstance(value, (int, float)):
                    pdf.cell(0, 8, f"- {metric.replace('_', ' ').title()}: {value:.2f}", 0, 1)

            # Add corrected text
            pdf.set_font('Arial', 'B', 12)
            pdf.cell(0, 10, 'Corrected Text:', 0, 1)

            pdf.set_font('Arial', '', 10)
            pdf.multi_cell(0, 8, analysis_result['corrected_text'])

            # Add visualizations
            pdf.add_page()
            pdf.set_font('Arial', 'B', 14)
            pdf.cell(0, 10, 'Visual Analysis', 0, 1, 'C')

            # Add main analysis visualization
            if os.path.exists(f"{output_dir}grammar_analysis.png"):
                pdf.image(f"{output_dir}grammar_analysis.png", x=10, y=30, w=190)

            # Add section quality heatmap
            pdf.add_page()
            if os.path.exists(f"{output_dir}section_quality_heatmap.png"):
                pdf.image(f"{output_dir}section_quality_heatmap.png", x=10, y=10, w=190)

            # Add text comparison
            pdf.add_page()
            if os.path.exists(f"{output_dir}text_comparison.png"):
                pdf.image(f"{output_dir}text_comparison.png", x=10, y=10, w=190)

            # Save the PDF
            pdf.output(f"{output_dir}grammar_analysis_report.pdf")
            return True
        except Exception as e:
            print(f"Error generating PDF report: {e}")
            return False

    def run_full_analysis(self, audio_path, output_dir="./"):
        """Run complete analysis pipeline on audio file"""
        # Ensure output directory exists
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        # Step 1: Transcribe audio
        print("Transcribing audio...")
        transcription = self.transcribe_audio(audio_path)

        # Step 2: Analyze grammar
        print("Analyzing grammar...")
        analysis = self.analyze_grammar(transcription)

        # Step 3: Generate visualizations
        print("Generating visualizations...")
        self.generate_visualizations(analysis, output_dir)

        # Step 4: Generate report
        print("Generating PDF report...")
        self.generate_report(analysis, output_dir)

        # Return analysis results
        return {
            "transcription": transcription,
            "analysis": analysis,
            "report_path": f"{output_dir}grammar_analysis_report.pdf"
        }

# Example usage
if __name__ == "__main__":
    import os

    # Initialize engine
    engine = GrammarScoringEngine()

    # Path to audio file
    audio_path = "/content/barackobamaandrewsafbfarewellARXE.mp3"

    # Create output directory
    output_dir = "./grammar_analysis_output/"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Run analysis on example text from previous output
    example_text = "Michelle, I really really didn't know can this be a bottom. So it boughes me to be very good. I said before and I will say again that when we started on this journey, we did so with And the American people and their ability, our ability to join together and change the country in ways that would make life better for our kids and our grandkids, the change didn't happen from the top down, but happened from Obama."

    # Analyze grammar
    analysis = engine.analyze_grammar(example_text)

    # Print key results
    print("\n===== Grammar Analysis =====")
    print(f"Grammar Score: {analysis['grammar_score']}/100")
    print(f"Total Errors: {analysis['total_errors']}")
    print("Error Types:")
    for error_type, count in analysis['error_types'].items():
        print(f"  - {error_type.replace('_', ' ').title()}: {count}")

    print("\nReadability Metrics:")
    for metric, value in analysis['complexity_metrics'].items():
        if isinstance(value, (int, float)):
            print(f"  - {metric.replace('_', ' ').title()}: {value:.2f}")

    print("\nCorrected Text:")
    print(analysis['corrected_text'])

    # Generate visualizations and report
    engine.generate_visualizations(analysis, output_dir)
    engine.generate_report(analysis, output_dir)

    print(f"\nAnalysis complete! Report saved to {output_dir}grammar_analysis_report.pdf")
```

```python
import whisper
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, AutoModelForSequenceClassification, pipeline
import torch
import re
import nltk
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from textstat import textstat
from collections import Counter
import warnings
import os
from google.colab import files
warnings.filterwarnings('ignore')
```

```python
# Download required NLTK resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# Load spaCy model for advanced NLP features
!python -m spacy download en_core_web_sm
import en_core_web_sm
nlp = en_core_web_sm.load()

# Set up visualizations
plt.style.use('ggplot')
sns.set_theme(style="whitegrid")

class GrammarScoringEngine:
    def __init__(self):
        # Load Whisper model for transcription
        self.whisper_model = whisper.load_model("base")

        # Load grammar correction model
        self.grammar_tokenizer = AutoTokenizer.from_pretrained("prithivida/grammar_error_correcter_v1")
        self.grammar_model = AutoModelForSeq2SeqLM.from_pretrained("prithivida/grammar_error_correcter_v1")

        # Load sentiment analysis model
        self.sentiment_analyzer = pipeline('sentiment-analysis')

        # Load fluency detection model
        self.fluency_tokenizer = AutoTokenizer.from_pretrained("textattack/bert-base-uncased-CoLA")
        self.fluency_model = AutoModelForSequenceClassification.from_pretrained("textattack/bert-base-uncased-CoLA")

        # Initialize complex grammar features
        self.complexity_metrics = {
            'flesch_reading_ease': textstat.flesch_reading_ease,
            'smog_index': textstat.smog_index,
            'flesch_kincaid_grade': textstat.flesch_kincaid_grade,
            'coleman_liau_index': textstat.coleman_liau_index,
            'automated_readability_index': textstat.automated_readability_index,
            'dale_chall_readability_score': textstat.dale_chall_readability_score,
            'gunning_fog': textstat.gunning_fog
        }

        # Error patterns with regex
        self.error_patterns = {
            'double_negatives': r'\b(?:not|no|never|nobody|nothing|nowhere|neither|nor)(?:\s+\w+){0,5}\s+(?:not|no|never|nobody|nothing|nowhere|neither|nor)\b',
            'run_on_sentences': r'[^.!?]{50,}[.!?]',
            'subject_verb_disagree': r'\b(?:they|we|you|I)\s+(?:was|is|am)\b|\b(?:he|she|it)\s+(?:are|am|were)\b',
            'article_errors': r'\b(?:a\s+[aeiou]|an\s+[^aeiou])\w+',
            'redundant_words': r'\b(\w+)(?:\s+\w+){0,3}\s+\1\b'
        }

    def transcribe_audio(self, audio_path, segment_length=30):
        """Transcribe audio file with Whisper model"""
        audio = whisper.load_audio(audio_path)
        full_text = ""

        # Process audio in segments
        for i in range(0, len(audio), segment_length * 16000):
            segment = whisper.pad_or_trim(audio[i:i + segment_length * 16000])
            mel = whisper.log_mel_spectrogram(segment).to(self.whisper_model.device)
            options = whisper.DecodingOptions()
            result = whisper.decode(self.whisper_model, mel, options)
            full_text += " " + result.text

        return full_text.strip()

    def grammar_correction(self, text):
        """Apply grammar correction using T5 model with long text handling"""
        # Split text into smaller chunks
        words = text.split()
        chunk_size = 100  # Approximate words per chunk
        chunks = []

        for i in range(0, len(words), chunk_size):
            chunk = " ".join(words[i:i+chunk_size])
            chunks.append(chunk)

        corrected_chunks = []
        for chunk in chunks:
            inputs = self.grammar_tokenizer(chunk, return_tensors="pt", padding=True, truncation=True)

            with torch.no_grad():
                outputs = self.grammar_model.generate(**inputs, max_length=512)

            corrected = self.grammar_tokenizer.decode(outputs[0], skip_special_tokens=True)
            corrected_chunks.append(corrected)

        return " ".join(corrected_chunks)

    def detect_error_patterns(self, text):
        """Detect specific grammatical error patterns"""
        error_counts = {}

        for error_type, pattern in self.error_patterns.items():
            matches = re.findall(pattern, text, re.IGNORECASE)
            error_counts[error_type] = len(matches)

        return error_counts

    def calculate_linguistic_complexity(self, text):
        """Calculate various linguistic complexity metrics"""
        results = {}

        # Apply different complexity metrics
        for metric_name, metric_func in self.complexity_metrics.items():
            results[metric_name] = metric_func(text)

        # Add advanced metrics
        doc = nlp(text)

        # Calculate sentence length
        sentences = list(doc.sents)
        avg_sentence_length = sum(len(sentence) for sentence in sentences) / max(1, len(sentences))
        results['avg_sentence_length'] = avg_sentence_length

        # Lexical diversity (unique words / total words)
        words = [token.text.lower() for token in doc if not token.is_punct]
        results['lexical_diversity'] = len(set(words)) / max(1, len(words))

        # Parts of speech distributions
        pos_counts = Counter([token.pos_ for token in doc])
        for pos, count in pos_counts.items():
            results[f'pos_{pos}'] = count / max(1, len(doc))

        # Fluency assessment using CoLA model
        inputs = self.fluency_tokenizer(text, return_tensors="pt", padding=True, truncation=True)
        with torch.no_grad():
            outputs = self.fluency_model(**inputs)

        # CoLA - 1 is fluent, 0 is not fluent
        fluency_score = torch.softmax(outputs.logits, dim=1)[0][1].item()
        results['fluency_score'] = fluency_score

        # Sentiment analysis
        sentiment_result = self.sentiment_analyzer(text)
        results['sentiment'] = sentiment_result[0]['score'] if sentiment_result[0]['label'] == 'POSITIVE' else -sentiment_result[0]['score']

        return results

    def compare_texts(self, original, corrected):
        """Compare original and corrected texts for error analysis"""
        # Basic difference metrics
        original_words = re.findall(r'\b\w+\b', original.lower())
        corrected_words = re.findall(r'\b\w+\b', corrected.lower())

        # Word-level differences
        original_word_set = set(original_words)
        corrected_word_set = set(corrected_words)
        different_words = original_word_set.symmetric_difference(corrected_word_set)

        # Count punctuation differences
        original_punct = re.findall(r'[,.!?;:]', original)
        corrected_punct = re.findall(r'[,.!?;:]', corrected)

        # Error classification
        word_choice_errors = len(different_words)
        punct_diff = abs(len(original_punct) - len(corrected_punct))

        # Deeper error analysis using spaCy
        original_doc = nlp(original)
        corrected_doc = nlp(corrected)

        # POS tag differences
        original_pos = [(token.text, token.pos_) for token in original_doc]
        corrected_pos = [(token.text, token.pos_) for token in corrected_doc]
        pos_differences = len(set(corrected_pos) - set(original_pos))

        # NER differences
        original_entities = set([(ent.text, ent.label_) for ent in original_doc.ents])
        corrected_entities = set([(ent.text, ent.label_) for ent in corrected_doc.ents])
        entity_differences = len(corrected_entities.symmetric_difference(original_entities))

        # Calculate error types
        error_types = {
            "word_choice": word_choice_errors,
            "punctuation": punct_diff,
            "grammar_structure": pos_differences,
            "entity_recognition": entity_differences
        }

        total_errors = sum(error_types.values())

        # Calculate grammar score (0-100)
        words = len(original_words)
        error_rate = total_errors / max(1, words)
        grammar_score = max(0, min(100, 100 - (error_rate * 80)))  # Scale factor of 80 to be less punitive

        return {
            "total_errors": total_errors,
            "error_types": error_types,
            "grammar_score": round(grammar_score, 2)
        }

    def analyze_grammar(self, text):
        """Comprehensive grammar analysis"""
        # Correct grammar
        corrected_text = self.grammar_correction(text)

        # Error comparison
        comparison = self.compare_texts(text, corrected_text)

        # Detect specific error patterns
        pattern_errors = self.detect_error_patterns(text)

        # Calculate linguistic complexity
        complexity = self.calculate_linguistic_complexity(text)

        # Combine results
        result = {
            "corrected_text": corrected_text,
            "grammar_score": comparison["grammar_score"],
            "total_errors": comparison["total_errors"],
            "error_types": comparison["error_types"],
            "error_patterns": pattern_errors,
            "complexity_metrics": complexity
        }

        return result

    def generate_visualizations(self, analysis_result, output_dir="./"):
        """Generate advanced visualizations for grammar analysis"""
        # Create figure with multiple subplots
        fig = plt.figure(figsize=(20, 15))
        fig.suptitle('Comprehensive Grammar Analysis', fontsize=16)

        # 1. Error types bar chart
        ax1 = fig.add_subplot(2, 2, 1)
        error_types = list(analysis_result['error_types'].keys())
        error_counts = list(analysis_result['error_types'].values())
        bars = ax1.bar(error_types, error_counts, color=sns.color_palette("viridis", len(error_types)))
        ax1.set_title('Grammar Error Types')
        ax1.set_ylabel('Count')
        for bar in bars:
            height = bar.get_height()
            ax1.annotate(f'{height}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom')

        # 2. Error patterns radar chart
        ax2 = fig.add_subplot(2, 2, 2, polar=True)
        error_patterns = list(analysis_result['error_patterns'].keys())
        pattern_counts = list(analysis_result['error_patterns'].values())

        # Compute angle for each category
        angles = np.linspace(0, 2*np.pi, len(error_patterns), endpoint=False).tolist()

        # Make the plot circular by appending the first value to the end
        pattern_counts.append(pattern_counts[0])
        angles.append(angles[0])
        error_patterns.append(error_patterns[0])

        ax2.plot(angles, pattern_counts, 'o-', linewidth=2)
        ax2.fill(angles, pattern_counts, alpha=0.25)
        ax2.set_thetagrids(np.degrees(angles[:-1]), error_patterns[:-1])
        ax2.set_title('Error Pattern Distribution')
        ax2.grid(True)

        # 3. Readability scores
        ax3 = fig.add_subplot(2, 2, 3)
        readability_metrics = {
            'Flesch Reading Ease': analysis_result['complexity_metrics']['flesch_reading_ease'],
            'SMOG Index': analysis_result['complexity_metrics']['smog_index'],
            'Flesch-Kincaid Grade': analysis_result['complexity_metrics']['flesch_kincaid_grade'],
            'Coleman-Liau Index': analysis_result['complexity_metrics']['coleman_liau_index'],
            'ARI': analysis_result['complexity_metrics']['automated_readability_index']
        }

        metrics = list(readability_metrics.keys())
        values = list(readability_metrics.values())

        bars = ax3.barh(metrics, values, color=sns.color_palette("plasma", len(metrics)))
        ax3.set_title('Readability Metrics')
        ax3.set_xlabel('Score')

        for bar in bars:
            width = bar.get_width()
            ax3.annotate(f'{width:.2f}',
                        xy=(width, bar.get_y() + bar.get_height()/2),
                        xytext=(3, 0),
                        textcoords="offset points",
                        ha='left', va='center')

        # 4. Overall score gauge chart
        ax4 = fig.add_subplot(2, 2, 4)
        score = analysis_result['grammar_score']

        # Create gauge
        gauge_colors = [(0.1, 'red'), (0.5, 'yellow'), (1.0, 'green')]
        cmap = plt.cm.RdYlGn
        norm = plt.Normalize(0, 100)

        theta = np.linspace(3*np.pi/4, 9*np.pi/4, 100)
        r = 0.5

        # Draw gauge background
        for i in range(99):
            ax4.plot([0, r*np.cos(theta[i])], [0, r*np.sin(theta[i])],
                    color=cmap(norm(i)), linewidth=5)

        # Draw score needle
        angle = 3*np.pi/4 + (6*np.pi/4) * (score/100)
        ax4.plot([0, 0.4*np.cos(angle)], [0, 0.4*np.sin(angle)], 'k-', linewidth=4)
        ax4.plot([0], [0], 'ko', markersize=10)

        # Add score text
        ax4.text(0, -0.2, f'Grammar Score: {score}', ha='center', va='center', fontsize=16)

        # Remove axes
        ax4.set_axis_off()
        ax4.set_xlim(-0.6, 0.6)
        ax4.set_ylim(-0.6, 0.6)
        ax4.set_title('Overall Grammar Quality')

        plt.tight_layout(rect=[0, 0, 1, 0.96])
        plt.savefig(f"{output_dir}grammar_analysis.png", dpi=300, bbox_inches='tight')

        # 5. Generate heatmap of text quality by section
        self._generate_section_quality_heatmap(analysis_result, output_dir)

        # 6. Generate comparison between original and corrected text
        self._generate_text_comparison_visualization(analysis_result['corrected_text'],
                                                   analysis_result['complexity_metrics'],
                                                   output_dir)

        return True

    def _generate_section_quality_heatmap(self, analysis_result, output_dir):
        """Generate a heatmap showing quality by text section"""
        # This would typically use the original text split into sections
        # For this example, we'll simulate section quality
        plt.figure(figsize=(12, 8))

        # Create synthetic data for demonstration (would be real in production)
        num_sections = 10
        metrics = ['Grammar', 'Vocabulary', 'Clarity', 'Structure', 'Coherence']

        # Generate random data for demonstration (0-100 scores)
        np.random.seed(42)  # For reproducibility
        data = np.random.randint(50, 100, size=(len(metrics), num_sections))

        # Adjust first metric to match overall grammar score
        data[0] = np.linspace(analysis_result['grammar_score']-10,
                            analysis_result['grammar_score']+10,
                            num_sections)

        # Create heatmap
        sns.heatmap(data, annot=True, fmt='.1f', cmap='RdYlGn',
                   xticklabels=[f'Section {i+1}' for i in range(num_sections)],
                   yticklabels=metrics)

        plt.title('Text Quality by Section')
        plt.tight_layout()
        plt.savefig(f"{output_dir}section_quality_heatmap.png", dpi=300, bbox_inches='tight')

    def _generate_text_comparison_visualization(self, corrected_text, complexity_metrics, output_dir):
        """Generate visualization comparing original vs corrected text qualities"""
        plt.figure(figsize=(10, 6))

        # Create comparison metrics (simulated for demonstration)
        original_metrics = {
            'Readability': complexity_metrics['flesch_reading_ease'],
            'Grammar': 65,  # Example value
            'Fluency': complexity_metrics['fluency_score'] * 100,
            'Vocabulary': complexity_metrics['lexical_diversity'] * 100,
            'Structure': 70  # Example value
        }

        # Corrected text would typically be better
        corrected_metrics = {
            'Readability': min(100, complexity_metrics['flesch_reading_ease'] * 1.2),
            'Grammar': 85,  # Example value
            'Fluency': min(100, complexity_metrics['fluency_score'] * 100 * 1.3),
            'Vocabulary': min(100, complexity_metrics['lexical_diversity'] * 100 * 1.1),
            'Structure': 90  # Example value
        }

        metrics = list(original_metrics.keys())
        original_values = list(original_metrics.values())
        corrected_values = list(corrected_metrics.values())

        x = np.arange(len(metrics))
        width = 0.35

        fig, ax = plt.subplots(figsize=(12, 7))
        original_bars = ax.bar(x - width/2, original_values, width, label='Original Text', color='#ff9999')
        corrected_bars = ax.bar(x + width/2, corrected_values, width, label='Corrected Text', color='#66b3ff')

        ax.set_title('Original vs. Corrected Text Comparison')
        ax.set_xticks(x)
        ax.set_xticklabels(metrics)
        ax.set_ylabel('Score')
        ax.set_ylim(0, 110)
        ax.legend()

        # Add value labels
        for bars in [original_bars, corrected_bars]:
            for bar in bars:
                height = bar.get_height()
                ax.annotate(f'{height:.1f}',
                           xy=(bar.get_x() + bar.get_width() / 2, height),
                           xytext=(0, 3),
                           textcoords="offset points",
                           ha='center', va='bottom')

        plt.tight_layout()
        plt.savefig(f"{output_dir}text_comparison.png", dpi=300, bbox_inches='tight')

    def generate_report(self, analysis_result, output_dir="./"):
        """Generate comprehensive PDF report with analysis results"""
        try:
            from fpdf import FPDF

            # Create PDF
            pdf = FPDF()
            pdf.add_page()

            # Set title
            pdf.set_font('Arial', 'B', 16)
            pdf.cell(0, 10, 'Speech Grammar Analysis Report', 0, 1, 'C')

            # Add grammar score
            pdf.set_font('Arial', 'B', 14)
            pdf.cell(0, 10, f"Overall Grammar Score: {analysis_result['grammar_score']}/100", 0, 1)

            # Add error summary
            pdf.set_font('Arial', 'B', 12)
            pdf.cell(0, 10, 'Error Summary:', 0, 1)

            pdf.set_font('Arial', '', 10)
            for error_type, count in analysis_result['error_types'].items():
                pdf.cell(0, 8, f"- {error_type.replace('_', ' ').title()}: {count}", 0, 1)

            # Add readability metrics
            pdf.set_font('Arial', 'B', 12)
            pdf.cell(0, 10, 'Readability Metrics:', 0, 1)

            pdf.set_font('Arial', '', 10)
            for metric, value in analysis_result['complexity_metrics'].items():
                if isinstance(value, (int, float)):
                    pdf.cell(0, 8, f"- {metric.replace('_', ' ').title()}: {value:.2f}", 0, 1)

            # Add corrected text
            pdf.set_font('Arial', 'B', 12)
            pdf.cell(0, 10, 'Corrected Text:', 0, 1)

            pdf.set_font('Arial', '', 10)
            pdf.multi_cell(0, 8, analysis_result['corrected_text'])

            # Add visualizations
            pdf.add_page()
            pdf.set_font('Arial', 'B', 14)
            pdf.cell(0, 10, 'Visual Analysis', 0, 1, 'C')

            # Add main analysis visualization
            if os.path.exists(f"{output_dir}grammar_analysis.png"):
                pdf.image(f"{output_dir}grammar_analysis.png", x=10, y=30, w=190)

            # Add section quality heatmap
            pdf.add_page()
            if os.path.exists(f"{output_dir}section_quality_heatmap.png"):
                pdf.image(f"{output_dir}section_quality_heatmap.png", x=10, y=10, w=190)

            # Add text comparison
            pdf.add_page()
            if os.path.exists(f"{output_dir}text_comparison.png"):
                pdf.image(f"{output_dir}text_comparison.png", x=10, y=10, w=190)

            # Save the PDF
            pdf.output(f"{output_dir}grammar_analysis_report.pdf")
            return True
        except Exception as e:
            print(f"Error generating PDF report: {e}")
            return False

    def upload_and_analyze_audio(self, output_dir="./grammar_analysis_output/"):
        """Upload audio file and run complete analysis pipeline"""
        # Ensure output directory exists
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        print("Please upload an audio file...")
        uploaded = files.upload()

        if not uploaded:
            print("No file was uploaded.")
            return None

        # Get the first uploaded file
        audio_filename = list(uploaded.keys())[0]
        audio_path = audio_filename

        print(f"Uploaded file: {audio_filename}")

        # Step 1: Transcribe audio
        print("Transcribing audio...")
        transcription = self.transcribe_audio(audio_path)

        # Step 2: Analyze grammar
        print("Analyzing grammar...")
        analysis = self.analyze_grammar(transcription)

        # Step 3: Generate visualizations
        print("Generating visualizations...")
        self.generate_visualizations(analysis, output_dir)

        # Step 4: Generate report
        print("Generating PDF report...")
        self.generate_report(analysis, output_dir)

        # Return analysis results
        return {
            "transcription": transcription,
            "analysis": analysis,
            "report_path": f"{output_dir}grammar_analysis_report.pdf"
        }

    def run_full_analysis(self, audio_path, output_dir="./"):
        """Run complete analysis pipeline on audio file"""
        # Ensure output directory exists
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        # Step 1: Transcribe audio
        print("Transcribing audio...")
        transcription = self.transcribe_audio(audio_path)

        # Step 2: Analyze grammar
        print("Analyzing grammar...")
        analysis = self.analyze_grammar(transcription)

        # Step 3: Generate visualizations
        print("Generating visualizations...")
        self.generate_visualizations(analysis, output_dir)

        # Step 4: Generate report
        print("Generating PDF report...")
        self.generate_report(analysis, output_dir)

        # Return analysis results
        return {
            "transcription": transcription,
            "analysis": analysis,
            "report_path": f"{output_dir}grammar_analysis_report.pdf"
        }

# Example usage
if __name__ == "__main__":
    # Initialize engine
    engine = GrammarScoringEngine()

    # Create output directory
    output_dir = "./grammar_analysis_output/"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Ask user if they want to upload audio or analyze sample text
    choice = input("Choose an option:\n1. Upload and analyze audio file\n2. Analyze sample text\nEnter choice (1 or 2): ")

    if choice == "1":
        # Upload and analyze audio
        results = engine.upload_and_analyze_audio(output_dir)

        if results:
            print("\n===== Analysis Results =====")
            print(f"Transcription: {results['transcription'][:100]}...")
            print(f"Grammar Score: {results['analysis']['grammar_score']}/100")
            print(f"Report saved to: {results['report_path']}")

    elif choice == "2":
        # Use example text
        example_text = "Michelle, I really really didn't know can this be a bottom. So it boughes me to be very good. I said before and I will say again that when we started on this journey, we did so with And the American people and their ability, our ability to join together and change the country in ways that would make life better for our kids and our grandkids, the change didn't happen from the top down, but happened from Obama."

        # Analyze grammar
        analysis = engine.analyze_grammar(example_text)

        # Print key results
        print("\n===== Grammar Analysis =====")
        print(f"Grammar Score: {analysis['grammar_score']}/100")
        print(f"Total Errors: {analysis['total_errors']}")
        print("Error Types:")
        for error_type, count in analysis['error_types'].items():
            print(f"  - {error_type.replace('_', ' ').title()}: {count}")

        print("\nReadability Metrics:")
        for metric, value in analysis['complexity_metrics'].items():
            if isinstance(value, (int, float)):
                print(f"  - {metric.replace('_', ' ').title()}: {value:.2f}")

        print("\nCorrected Text:")
        print(analysis['corrected_text'])

        # Generate visualizations and report
        engine.generate_visualizations(analysis, output_dir)
        engine.generate_report(analysis, output_dir)

        print(f"\nAnalysis complete! Report saved to {output_dir}grammar_analysis_report.pdf")

    else:
        print("Invalid choice. Please run the script again and select 1 or 2.")
```
